```python
def six_plots(res, true_val, names=None, compare_name="параметров"):

    f, (ax1, ax2) = plt.subplots(1, 2, sharey=False, figsize=(17,6))
    for i in range(len(res)):
        ax1.plot(res[i][2], [np.log(abs(x[0] - true_val)) for x in res[i][1]])                  
    ax1.legend(names)
    ax1.set_ylabel("log(|L(w) - L(w*)|)")
    ax1.set_xlabel("iteration")
    ax1.set_title("Сходимость от итераций для разных " + compare_name)
    ax1.grid(which='both',
            color = 'gray',
            linestyle = ':')
  
    for i in range(len(res)):
        ax2.plot(res[i][2], np.log10(np.abs(res[i][5])))
                    
    ax2.legend(names)
    ax2.set_ylabel("log(|grad(L(w)) / grad(L(w_0)|)")
    ax2.set_xlabel("iteration")
    ax2.set_title("Сходимость от итераций для разных " + compare_name)
    ax2.grid(which='both',
            color = 'gray',
            linestyle = ':')
    f.show()


    f, (ax1, ax2) = plt.subplots(1, 2, sharey=False,figsize=(17,6))
    for i in range(len(res)):
        ax1.plot(res[i][3], [np.log10(abs(x[0] - true_val)) for x in res[i][1]])                  
    ax1.legend(names)
    ax1.set_ylabel("log(|L(w) - L(w*)|)")
    ax1.set_xlabel("oracle calls")
    ax1.set_title("Сходимость от вызовов оракула для разных " + compare_name)
    ax1.grid(which='both',
            color = 'gray',
            linestyle = ':')
  
    for i in range(len(res)):
        ax2.plot(res[i][3], np.log10(np.abs(res[i][5])))
                    
    ax2.legend(names)
    ax2.set_ylabel("log(|grad(L(w)) / grad(L(w_0)|)")
    ax2.set_xlabel("oracle calls")
    ax2.set_title("Сходимость от вызовов оракула для разных " + compare_name)
    ax2.grid(which='both',
            color = 'gray',
            linestyle = ':')
    f.show()



    f, (ax1, ax2) = plt.subplots(1, 2, sharey=False,figsize=(17,6))
    for i in range(len(res)):
        ax1.plot(res[i][4], [np.log10(abs(x[0] - true_val)) for x in res[i][1]])                  
    ax1.legend(names)
    ax1.set_ylabel("log(|L(w) - L(w*)|)")
    ax1.set_xlabel("time, s")
    ax1.set_title("Сходимость от времени для разных " + compare_name)
    ax1.grid(which='both',
            color = 'gray',
            linestyle = ':')
  
    for i in range(len(res)):
        ax2.plot(res[i][4], np.log10(np.abs(res[i][5])))
                    
    ax2.legend(names)
    ax2.set_ylabel("log(|grad(L(w)) / grad(L(w_0)|)")
    ax2.set_xlabel("time, s")
    ax2.set_title("Сходимость от времени для разных " + compare_name)
    ax2.grid(which='both',
             color = 'gray',
             linestyle = ':')
    f.show()
```

```python
def lbfgs(g, x, S, Y, RO, gamma):
    m = len(Y)
    alpha = [0] * m
    q = g
    for i in range(m-1, -1, -1):
        alpha[i] = RO[i] * S[i].T @ q
        q = q - alpha[i] * Y[i]
    z = gamma * q
    for i in range(m):
        beta = RO[i] * Y[i].T @ z
        z += S[i] * (alpha[i] - beta)
    return -z

class lbfgs_optimize:
    def __init__(self):
        self.values = []
        self.iterations = []
        self.oracle_calls = []
        self.times = []
        self.grads = []
        
    def __call__(self, oracle, init_point, tolerance=1e-8, history_size=10, max_iter=1000):
        x = init_point
        
        time0 = time()
        v, g = oracle.fuse_value_grad(x)
        oracle_call = 1 
        g_start = g
        self.values.append(v)
        self.iterations.append(0)
        self.oracle_calls.append(oracle_call)
        self.grads.append(((g.T @ g) / (g_start.T @ g_start))[0][0])
        self.times.append(time() - time0) 
        
        # списки для хранения истории оптимизационного процесса
        S = []
        Y = []
        RO = []
        
        line_search = line_search_wolf()
        
        k = 0
        while ((g.T @ g) / (g_start.T @ g_start))[0][0] > tolerance:
            if k > max_iter:
                break
            if k == 0:
                gamma = 1
            else:
                gamma = S[-1].T @ Y[-1] / (Y[-1].T @ Y[-1])
                
            d = lbfgs(g, x, S, Y, RO, gamma)
            alpha, funcalls = line_search(oracle, x, d, stata=True)
            oracle_call += funcalls           
            
            x_prev = np.copy(x)
            g_prev = np.copy(g)
            x = x + alpha * d
            
            v, g = oracle.fuse_value_grad(x)
            oracle_call += 1    
            
            Y.append(g - g_prev)
            S.append(x - x_prev)
            RO.append(1 / (Y[-1].T @ S[-1]))
            if k > history_size:
                Y.pop(0)
                S.pop(0)
                RO.pop(0)  
                
            print('iteration: {}'.format(k + 1))
            print('value of CE: {}'.format(v[0][0]))
            print('norm_grad_sq:', (g.T @ g)[0][0])
            print('alpha', alpha.squeeze())
            print() 

            self.values.append(v)
            self.iterations.append(k + 1)
            self.oracle_calls.append(oracle_call)
            self.times.append(time() - time0)
            self.grads.append(((g.T @ g) / (g_start.T @ g_start))[0][0])
            
            k += 1

        return x
```

### Разные датасеты

```python
oracle1 = make_oracle("a1a.txt")
oracle2 = make_oracle("breast_cancer_scale.txt")

# my dataset
np.random.seed(37)
a, b = 2 * np.random.random(2) - 1
X_ = np.random.normal(loc=0.0, scale=1.0, size=1000).reshape(-1, 1) 
y = ((a * X_ + b) >= 0).astype(int)
X = np.hstack([X_, np.ones(X_.shape[0]).reshape(-1, 1)])
oracle3 = Oracle(X, y)

def true_val(oracle):
    clf = LogisticRegression(penalty='none',
                             fit_intercept=False,
                             max_iter=1000).fit(oracle.X, np.ravel(oracle.y))
    w_sk = clf.coef_[0].reshape(-1, 1)
    true_val = oracle.value(w_sk)[0][0]
    return true_val

true_val1 = true_val(oracle1)
true_val2 = true_val(oracle2)
true_val3 = true_val(oracle3)
```

```python
# a1a dataset
oracle = oracle1
history_sizes = [1, 5, 10, 20, 50, 100]

w0 = np.zeros(oracle.X.shape[1]).reshape(-1, 1)
# w0 = np.random.randn(oracle.X.shape[1]).reshape(-1, 1)
# w0 = np.ones(oracle.X.shape[1]).reshape(-1, 1)

res = []
for history_size in history_sizes:
    start = time()
    print("hist_size = {} + LBFGS: START".format(history_size))
    print()
    optimize = lbfgs_optimize()
    w_my = optimize(oracle, w0, history_size=history_size)

    vals = optimize.values
    iters = optimize.iterations
    or_calls = optimize.oracle_calls
    times = optimize.times
    grads = optimize.grads

    res.append([w_my, vals, iters, or_calls, times, grads])
    print("hist_size = {} + LBFGS: END".format(history_size))
    print("CE value:", oracle.value(w_my).squeeze())
    print("Total oracle calls: ", or_calls[-1])
    print("Time: ", time() - start, "sec")
    print("----------------------------------------------------------")
    print()
```

```python
six_plots(res, true_val1, names=list(map(lambda x: "history_size = " + str(x), history_sizes)),
          compare_name="history_size")
```

На графиках видно, что вначале все кривые ведут себя одинаково, это объясняется тем, что происходит процесс накопления истории и при достижении заданного размера начинаются расхождения кривых.

В эксперименте наблюдаем, что с увеличением history_size время работы и количество итераций оптимизационного процесса уменьшается, то есть чем больше памяти используем, тем быстрее сходимся. Для размеров 50, 100 потребовалось 100 итераций, а для размеров 10, 20 -- 150 - 250. Для размеров начиная с 10 потребовалось не более 700 вызовов оракула. Сходимость получилась линейная.

Все методы отработали менее чем за 1.5 секунды. С большим размером истории -- около 0.5 сек.

### Сравнение с другими методами

```python
oracle = oracle1
w0 = np.zeros(oracle.X.shape[1]).reshape(-1, 1)
# w0 = np.random.randn(oracle.X.shape[1]).reshape(-1, 1)
# w0 = np.ones(oracle.X.shape[1]).reshape(-1, 1)

methods = [optimize_gd(), optimize_newton(),
           hfn_optimize(), lbfgs_optimize()]

names = ["GD", "Newton", "HFN", "L-BFGS"]


res = []
for i in range(len(methods)):
    start = time()
    print("wolf + {}: START".format(names[i]))
    print()
    optimize = methods[i]
    if i == 3:
        w_my = optimize(oracle, w0, history_size=10)
    else:
        line_search = line_search_wolf()
        # print(w.shape)
        w_my = optimize(oracle, w0, line_search)
    
    vals = optimize.values
    iters = optimize.iterations
    or_calls = optimize.oracle_calls
    times = optimize.times
    grads = optimize.grads
    
    res.append([w_my, vals, iters, or_calls, times, grads])
    print("wolf + {}: END".format(names[i]))
    print("CE value:", oracle.value(w_my).squeeze())
    print("Total oracle calls: ", or_calls[-1])
    print("Time: ", time() - start, "sec")
    print("----------------------------------------------------------")
    print()
```

```python
six_plots(res[1:], true_val1, names=names[1:], compare_name="методов")
```

```python
print("Информация про GD:")
print("Значение L в найденной оптимальной точке: {:1.5f}".format(res[0][1][-1][0][0]))
print("Итераций потребовалось: ", res[0][2][-1])
print("Вызовов оракула потребовалось: ", res[0][3][-1])
print("Времени в секундах заняло: {:1.3f}".format(res[0][4][-1]))
```

На графике не отображён GD, так как ему требуется значительно больше итераций/времени/вызовов оракула и графики были бы не информативны. Из графиков видно, что по итерациям быстрее всех оказались Ньютон и HFN, но не смотря на это L-BFGS показал лучшее время на равне с Ньютоном (1 сек), в то время как HFN сошёлся за 4 секунды. Это говорит о "дешёвых" вычислениях на каждой итерации в L-BFGS. При этом L-BFGS достиг точности 5 знаков после запятой.

Таким образом, был реализован метод оптимизации L-BFGS, позволяющий эффективно находить направление на каждой итерации без формирования матрицы Гессиана, что может сильно экономить память $O(nm)$.
